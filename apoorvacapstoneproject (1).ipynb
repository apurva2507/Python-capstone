```python
import pandas as pd
import numpy as np

# Creating Employee DataFrame
employee_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Name": ["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"],
    "Gender": ["M", "F", "M", "F", "F"],
    "City": ["Paris", "London", "Berlin", "Newyork", "Madrid"],
    "Age": [25, 27, 29, 31, 30]
}
employee_df = pd.DataFrame(employee_data)
employee_df.to_csv('employee.csv', index=False)
employee_df

# Creating Seniority Level DataFrame
seniority_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Designation Level": [2, 2, 3, 2, 3]
}
seniority_df = pd.DataFrame(seniority_data)
seniority_df.to_csv('seniority.csv', index=False)
seniority_df

# Creating Project DataFrame
project_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005", "A002", "A005", "A003", "A001", "A003", "A001", "A004", "A004", "A005"],
    "Project": ["Project 1", "Project 2", "Project 3", "Project 4", "Project 5", "Project 6", "Project 7", "Project 8", "Project 9", "Project 10", "Project 11", "Project 12", "Project 13", "Project 14"],
    "Cost": [1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000],
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", "Ongoing", "Finished", "Failed", "Ongoing", "Finished", "Finished"]
}
project_df = pd.DataFrame(project_data)
project_df.to_csv('project.csv', index=False)
project_df


# Load DataFrames from CSV
employee_df = pd.read_csv('employee.csv')
seniority_df = pd.read_csv('seniority.csv')
project_df = pd.read_csv('project.csv')

# Merge DataFrames 
final_df = pd.merge(employee_df, seniority_df, on='ID')
final_df = pd.merge(final_df, project_df, on='ID')

# Save Final DataFrame
final_df.to_csv('final.csv', index=False)

# Display the final DataFrame
print(final_df)
final_df
```

```python
## task 2:   

import pandas as pd

# Load project DataFrame from CSV
project_df = pd.read_csv('project.csv')

# Compute running average for missing values
for i in range(len(project_df)):
    if pd.isnull(project_df.at[i, 'Cost']):
        # Compute running average excluding the current missing value
        running_average = project_df['Cost'][:i].dropna().mean()
        project_df.at[i, 'Cost'] = running_average

# Save updated DataFrame
project_df.to_csv('project.csv', index=False)

# Display the updated DataFrame
print(project_df)
project_df
```

```python
# task:3
import pandas as pd

# Load employee DataFrame from CSV
employee_df = pd.read_csv('employee.csv')

# Print the original DataFrame to verify it was loaded correctly
print("Original DataFrame:")
print(employee_df)

# Verify column names
print("\nColumn Names:")
print(employee_df.columns)

# Strip any leading or trailing whitespace from the column names
employee_df.columns = employee_df.columns.str.strip()

# Print the cleaned column names
print("\nCleaned Column Names:")
print(employee_df.columns)

# Split "Name" column into "First Name" and "Last Name"
if 'Name' in employee_df.columns:
    employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(n=1, expand=True)

    # Remove the original "Name" column
    employee_df = employee_df.drop(columns=['Name'])

    # Print the updated DataFrame to verify the changes
    print("\nUpdated DataFrame:")
    print(employee_df)

    # Save updated DataFrame to CSV
    employee_df.to_csv('employee.csv', index=False)
else:
    print("\nError: 'Name' column not found in the DataFrame.")
    employee_df
```

```python
# task 5:
import pandas as pd

# Load the Final DataFrame from CSV
final_df = pd.read_csv('final.csv')

# Add a 5% bonus for finished projects
final_df['Bonus'] = final_df.apply(lambda row: row['Cost'] * 0.05 if row['Status'] == 'Finished' else 0, axis=1)

# Save the updated Final DataFrame to CSV
final_df.to_csv('final.csv', index=False)

# Display the updated Final DataFrame
print("Updated Final DataFrame with Bonus column:")
print(final_df)
final_df
```

```python
#task 6:
import pandas as pd

# Load the Final DataFrame from CSV
final_df = pd.read_csv('final.csv')

# Demote the designation level by 1 for failed projects
final_df.loc[final_df['Status'] == 'Failed', 'Designation Level'] += 1

# Remove employees whose designation level is above 4
final_df = final_df[final_df['Designation Level'] <= 4]

# Save the updated Final DataFrame to CSV
final_df.to_csv('final.csv', index=False)

# Display the updated Final DataFrame
print("Updated Final DataFrame after demotion and removal:")
print(final_df)
final_df
```

```python
#task 7:

import pandas as pd

# Load the Final DataFrame from CSV
final_df = pd.read_csv('final.csv')

# Print the column names to verify if 'First Name' and 'Gender' exist
print("Column Names in Final DataFrame:")
print(final_df.columns)

# Print a sample of the DataFrame to understand its current state
print("\nSample of the Final DataFrame:")
print(final_df.head())

# Check if 'First Name' and 'Gender' columns exist
if 'First Name' in final_df.columns and 'Gender' in final_df.columns:
    # Add "Mr." or "Mrs." to the first name based on gender
    final_df['First Name'] = final_df.apply(lambda row: 'Mr. ' + row['First Name'] if row['Gender'] == 'M' else 'Mrs. ' + row['First Name'], axis=1)

    # Drop the gender column
    final_df = final_df.drop(columns=['Gender'])

    # Save the updated Final DataFrame to CSV
    final_df.to_csv('final.csv', index=False)

    # Display the updated Final DataFrame in a tabular form
    print("\nUpdated Final DataFrame with titles and without Gender column:")
    print(final_df.to_string(index=False))
else:
    if 'First Name' not in final_df.columns:
        print("\nError: 'First Name' column not found in the DataFrame.")
    if 'Gender' not in final_df.columns:
        print("\nError: 'Gender' column not found in the DataFrame.")
        final_df
```

```python
#task 8:
# Promote designation level by 1 for employees older than 29
final_df.loc[final_df['Age'] > 29, 'Designation Level'] -= 1

# Save updated Final DataFrame
final_df.to_csv('final.csv', index=False)
final_df
```

```python
# task 9:
# Calculate total project cost for each employee
total_proj_cost = final_df.groupby('ID').agg({
    'First Name': 'first',
    'Cost': 'sum'
}).reset_index().rename(columns={'Cost': 'Total Cost'})

# Save TotalProjCost DataFrame
total_proj_cost.to_csv('total_proj_cost.csv', index=False)
total_proj_cost
```
